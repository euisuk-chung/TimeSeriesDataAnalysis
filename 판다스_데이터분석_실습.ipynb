### DataFrame 객체

DataFrame 객체는 행과 열로 이루어진 2차원 데이터를 다루기 위한 객체입니다. 열은 각각의 변수를 나타내고, 행은 각각의 관측치를 나타냅니다. DataFrame 객체는 여러 가지 방법으로 생성할 수 있습니다.


#### 1. DataFrame 생성하기

````python
# 리스트를 사용하여 DataFrame 객체 생성하기
import pandas as pd

data = [['A', 1], ['B', 2], ['C', 3]]
df = pd.DataFrame(data, columns=['col1', 'col2'])
print(df)
````

````python
# 딕셔너리를 사용하여 DataFrame 객체 생성하기
data = {'col1': ['A', 'B', 'C'], 'col2': [1, 2, 3]}
df = pd.DataFrame(data)
print(df)
````

````python
# (참고) CSV 파일로 데이터프레임 저장하기
print(df)
df.to_csv('data.csv', index=False)

# CSV 파일을 사용하여 DataFrame 객체 생성하기
df = pd.read_csv('data.csv')
print(df)
````

### Series 객체

Series 객체는 인덱스와 값으로 이루어진 1차원 데이터를 다루기 위한 객체입니다. Series 객체는 DataFrame 객체에서 열을 선택하여 추출할 수 있습니다.


#### 1. Series 생성하기

Series 객체에서는 리스트 또는 딕셔너리를 사용하여 객체를 생성할 수 있습니다.

````python
# 리스트를 사용하여 Series 객체 생성하기
import pandas as pd

data = [1, 2, 3]
s = pd.Series(data, index=['a', 'b', 'c'])
print(type(s))
print(s)
````

````python
# 딕셔너리를 사용하여 Series 객체 생성하기
data = {'a': 1, 'b': 2, 'c': 3}
s = pd.Series(data)
print(type(s))
print(s)
````

#### 2. Series 선택하기

Series 객체에서는 인덱스나 위치를 이용하여 데이터를 선택할 수 있습니다.


````python
### 선택하기
import pandas as pd
data = [1, 2, 3]
s = pd.Series(data, index=['a', 'b', 'c'])

# 출력
print("원본 시리즈:")
print(s)
print("\n")

# 데이터 추가
print("0번째 인덱스 선택")
print(s[0])

# 라벨을 사용한 인덱싱
print("라벨 'a'에 해당하는 데이터:")
print(s['a'])
````

#### 3. Series 조작하기

Series 객체에서는 인덱스나 위치를 이용하여 데이터를 선택할 수 있습니다.

````python
### 조작하기
import pandas as pd
data = [1, 2, 3]
s = pd.Series(data, index=['a', 'b', 'c'])

# 출력
print("원본 시리즈:")
print(s)
print("\n")

# 데이터 추가
s['d'] = 4
print("데이터 d - 4 추가:")
print(s)
print("\n")

# 데이터 삭제
s = s.drop('a')
print("데이터 a 인덱스 삭제:")
print(s)
print("\n")

# 데이터 수정
s['b'] = 5
print("데이터 b 인덱스 수정:")
print(s)
````

````python
# 엑셀 파일로 저장하기
df.to_excel('california_housing_train.xlsx', index=False)
````

### Excel 파일 불러오기

Excel 파일은 스프레드시트 프로그램에서 생성한 데이터를 저장하는 파일 형식입니다. 판다스에서는 `read_excel()` 함수를 이용하여 Excel 파일을 불러올 수 있습니다.



````python
import pandas as pd

# Excel 파일 불러오기
df = pd.read_excel('california_housing_train.xlsx')
df.head()
````

# 데이터 필터링하기
조건에 맞는 데이터를 필터링할 수 있습니다.


````python
print('필터링 전 데이터 크기:')
print(len(df['housing_median_age']))
print('-------------')

filtered_data = df[df['housing_median_age'] > 20]
print('필터링 후 데이터 크기:')
print(len(filtered_data['housing_median_age']))
````

# 데이터 그룹화하기
